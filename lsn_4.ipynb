### 1.1. 
Скачать следующие данные: kc-house-data и laptop_price

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('kc-house-data.csv')
df.head()
```

### 1.2. 
Изучите стоимости недвижимости

```python
p_houses = df['price'].value_counts()
prices = p_houses.index
counts = p_houses.values
plt.figure(figsize=[10,4])
sns.lineplot(x = prices, y = counts)
plt.grid()
plt.title('Распределение недвижимости по стоимости')
plt.xlabel('Стоимость')
plt.ylabel('Количество')
plt.show()
```

### 1.3. 
Изучите распределение квадратуры жилой площади.

```python
plt.figure(figsize=[8,4])
plt.hist(df['sqft_living'], bins=50)
plt.xlabel('Квадратура жилой площади')
plt.ylabel('Количество')
plt.title('Распределение квадратуры жилой площади');
```

### 1.4. 
Изучите распределение года постройки

```python
df['yr_built'].describe()
```

```python
plt.figure(figsize=[6, 3])
sns.histplot(df['yr_built'])
plt.title('Распределение недвижимости по году постройки')
plt.xlabel('Год')
plt.ylabel('Кол-во')
plt.show()
```

## Условие 2: 2 задача

### 2.1. 
Изучите распределение домов от наличия вида на набережную


```python
data = df['waterfront'].value_counts()

plt.figure(figsize=(6, 4))
plt.pie(data, autopct='%1.1f%%')
plt.legend(['no', 'yes']);
```

```python
matrix = df.corr(numeric_only=True)
matrix = np.round(matrix, 2)
matrix[np.abs(matrix) < 0.3] = 0
matrix
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(matrix, annot=True, linewidths=.5, cmap='coolwarm');
```
