## Задача 1:
На складе лежат разные фрукты в разном количестве.
Нужно написать функцию, которая на вход принимает любое количество названий фруктов и их количество, а возвращает общее количество фруктов на складе

```python
fruits = {'Апельсин': 100, 'Дыня': 150, 'Груша': 2, 'Яблоко': 800, 'Дуриан': 230}
def smvls (dct: dict) -> int:
    return sum(dct.values())
smvls(fruits)
```

## Задача 2:
Дан список с затратами на рекламу. Но в данных есть ошибки, некоторые затраты имеют отрицательную величину. Удалите такие значения из списка и посчитайте суммарные затраты
<pre><code>[100, 125, -90, 345, 655, -1, 0, 200]</code></pre>
Используйте list comprehensions

```python
lst = [100, 125, -90, 345, 655, -1, 0, 200]
def remsum (lst: list) -> int:
    return sum([i if i>0 else 0 for i in lst])
remsum(lst)
```

### Задание 3_2:
Найдите выручку компании в зависимости от месяца
Для этого напишите функцию, которая на вход принимает список с датами и список с выручкой, а на выходе словарь, где ключи - это месяцы, а значения - это выручка.
Используйте аннотирование типов.

```python
date = ['2021-09-14', '2021-12-15', '2021-09-08', '2021-12-05', '2021-10-09', '2021-09-30', '2021-12-22', '2021-11-29', '2021-12-24', '2021-11-26', '2021-10-27', '2021-12-18', '2021-11-09', '2021-11-23', '2021-09-27', '2021-10-02', '2021-12-27', '2021-09-20', '2021-12-13', '2021-11-01', '2021-11-09', '2021-12-06', '2021-12-08', '2021-10-09', '2021-10-31', '2021-09-30', '2021-11-09', '2021-12-13', '2021-10-26', '2021-12-09']
price = [1270, 8413, 9028, 3703, 5739, 4095, 295, 4944, 5723, 3701, 4471, 651, 7037, 4274, 6275, 4988, 6930, 2971, 6592, 2004, 2822, 519, 3406, 2732, 5015, 2008, 316, 6333, 5700, 2887]
def revmon (dt: list, pr: list) -> dict:
    dct = {dt[i] : pr[i] for i in range(len(dt))}
    revenue = {}
    for key in dct:
        if key[5:7] in revenue:
            revenue[key[5:7]] += dct[key]
        else:
            revenue[key[5:7]] = dct[key]
    return dict(sorted(revenue.items()))

revmon(date,price)
```
